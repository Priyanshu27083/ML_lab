```python
#23/07
import pandas as pd
sales = pd.DataFrame({2014:[100.5,150.8,200.9,30000,40000],2015:[12000,18000,22000,30000,45000],2016:[20000,50000,70000,100000,125000],2017:[50000,60000,70000,80000,90000]},index = ['Madhu','Kusum','Kinshuk','Ankit','Shruti'])
print(sales)
print(sales.index)
print(sales.columns)
print(sales.describe())
print(sales.shape)
print(sales.size)
print(sales.head(2))
print(sales.iloc[:,:2])
print(sales.ndim)
print(sales.dtypes)
print(sales.values)
print(sales.tail(2))
```

```python
dict1 = {2018:[160000,110000,500000,340000]}
sales2 = pd.DataFrame(dict1, index = ['Madhu','Kusum','Kinshuk','Ankit'])
print(sales2)
print(sales2.empty)
```

```python
#Appending
appended = pd.concat([sales,sales2], axis = 1)
print(appended)
```

```python
appended.transpose()
```

```python
appended[2017]
```

```python
select = appended.loc[['Madhu','Ankit'],[2017,2018]]
print(select)
```

```python
#finding according to row and column using loc
appended.loc['Shruti'][2016]
```

```python
#Adding a new row use .loc[]
appended.loc['Sumeet'] = [196.2, 37800, 52000, 78438,8852]
print(appended)
```

```python
#29/07
#Deleting a column we use axis = 1
appended.drop(2014,axis = 1, inplace = True)
```

```python
appended
```

```python
#To delete a row we use axis = 0
appended.drop('Kinshuk',axis = 0, inplace = True)
appended
```

```python
#Rename rows use df.rename(index = {'to_replace': 'replaced_name'})
appended.rename(index = {'Ankit':'Vivan', 'Madhu':'Shailesh'} , inplace = True)
appended
```

```python
#Update a particular value
appended.loc['Shailesh',2018] = 100000
appended
```

```python
#Sending dataframe to csv file without row and column labels
appended.to_csv('SalesFigures.csv', index = False, header = False)
```

```python
SalesRetrieved = pd.read_csv('SalesFigures.csv', header=None)
SalesRetrieved.index = appended.index
SalesRetrieved.columns = appended.columns
print(SalesRetrieved)
```

```python
#Exercise 2 13 a
df1 = pd.DataFrame({
    'Item': ['TV', 'TV', 'TV', 'AC'],
    'Company': ['LG', 'VIDEOCON', 'LG', 'SONY'],
    'Rupees': [12000, 10000, 15000, 14000],
    'USD': [700, 650, 800, 750]
})
df1
```

```python
#b
df1.loc[4] = ['Bag','Skybags',6500,75]
print(df1)
```

```python
#c max price
max_price = df1[(df1['Company'] == 'LG') & (df1['Item'] == 'TV')]['Rupees'].max()
print("\nMaximum price of LG TV:", max_price)
```

```python
#sum of all products
sum_products = df1['Rupees'].sum()
print("\nSum of all products:", sum_products)
```

```python
#median of a particular company
sony = df1[df1['Company'] == 'SONY']['USD'].median()
print("\nMedian USD price of Sony products:", sony)
```
