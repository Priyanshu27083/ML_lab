```python
import pandas as pd
travel = pd.read_csv('/content/Travel details dataset.csv')
travel.head()
```

```python
#Constant features
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold

# Assuming 'travel' DataFrame is already loaded
print([col for col in travel.columns if travel[col].isnull().sum() > 0])
travel = travel.dropna()
print([col for col in travel.columns if travel[col].isnull().sum() > 0])
```

```python
travel.dropna().head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(travel.drop(labels = ['Traveler nationality'],axis = 1),travel['Traveler nationality'],test_size = 0.3,random_state = 0)
x_train.shape,x_test.shape
```

```python
#Keeping all the constant value columns in a list
constant_features = [features for features in x_train.columns if x_train[features].std == 0]
print(len(constant_features))
```

```python
#Dropping the constant values
x_train.drop(labels=constant_features, axis=1, inplace=True)
x_test.drop(labels=constant_features, axis=1, inplace=True)
x_train.shape, x_test.shape
```

```python
#Quasi Constant Feature
import numpy as np
from sklearn.model_selection import train_test_split

# First, split your data
x_train, x_test, y_train, y_test = train_test_split(
    travel.drop(labels=['Traveler nationality', 'Traveler name'], axis=1),  # drop name as it's not useful
    travel['Traveler nationality'],
    test_size=0.3,
    random_state=0
)

# Find quasi-constant features
quasi_constant_features = []

for feature in x_train.columns:
#get the predominent value for every feature in the dataset
  predominant = (x_train[feature].value_counts() / float(len(x_train))).sort_values(ascending = False).values[0]

#check if predominant>99.9%
  if predominant > 0.999:
    quasi_constant_features.append(feature)

print(len(quasi_constant_features))
#Dropping the quasi_constant_feature columns
x_train.drop(labels = quasi_constant_features, axis = 1, inplace=True)
x_test.drop(labels = quasi_constant_features, axis = 1, inplace = True)
print(x_train.shape, x_test.shape)
```

```python
#Duplicate Features
#Create an empty list to store duplicated features
duplicated_features = []

# Step 2: Compare each feature with the rest
for i in range(0, len(x_train.columns)):
    col_1 = x_train.columns[i]

    for col_2 in x_train.columns[i + 1:]:
        if x_train[col_1].equals(x_train[col_2]):
            duplicated_features.append(col_2)

# Step 3: Convert to a set to keep only unique duplicate feature names
duplicated_features = set(duplicated_features)

# Optional: Print number of duplicated columns
print(f"Number of duplicated features: {len(duplicated_features)}")

# Step 4: Drop them from train and test datasets
x_train.drop(labels=duplicated_features, axis=1, inplace=True)
x_test.drop(labels=duplicated_features, axis=1, inplace=True)

# Step 5: Print shape to confirm changes
print(x_train.shape, x_test.shape)
```
