```python
#23/07
import pandas as pd
sales = pd.DataFrame({2014:[100.5,150.8,200.9,30000,40000],2015:[12000,18000,22000,30000,45000],2016:[20000,50000,70000,100000,125000],2017:[50000,60000,70000,80000,90000]},index = ['A','B','C','D','E'])
print(sales)
print(sales.index)
print(sales.columns)
print(sales.describe())
print(sales.shape)
print(sales.size)
print(sales.head(2))
print(sales.iloc[:,:2])
```

```python
dict1 = {2018:[160000,110000,500000,340000,900000]}
sales2 = pd.DataFrame(dict1, index = ['A','B','C','D','E'])
print(sales2)
print(sales2.empty)
```

```python
#Appending
appended = pd.concat([sales,sales2], axis = 1)
print(appended)
```
